```python
import pandas as pd
import re
import pickle
```

```python
with open(r"/content/drive/MyDrive/Mencius_research/Mencius_research/[논문 작업용] 맹자절문 남은부분.txt", 'rt', encoding='cp949') as f:
    file = [t.replace('\n','') for t in f.readlines()]
    
```

```python
file
```

```python
p = re.compile(r'[^a-zA-Z0-9\s·]+')
preprocessed=[re.sub(p,'',t) for t in file if t!='']
preprocessed = [re.sub(r'\s{2,}','',t) for t in preprocessed if t !='']
preprocessed = [t.strip() for t in preprocessed]
preprocessed
```

```python
roman = [re.match(r'(V|I|X)?I{,5}X?V?',t) for t in preprocessed]
roman = sorted(list(set([t for t in [t.group() for t in roman] if t!='']))) #I, II, III, IV, V, ...
num_dot = [re.match(r'[0-9]+·[0-9]+',t) for t in preprocessed]
num_dot = sorted(list(set([t.group() for t in num_dot if t!=None]))) # 1·1, 1·2 , ...
subtitle = [re.match(r'^[A-Z][A-Za-z]+\s[A-Z][A-Za-z\s]+I{1,2}$',t) for t in preprocessed]
subtitle = sorted([t.group() for t in subtitle if t!=None]) #'Liang Hui Wang I','Liang Hui Wang II',...
```

```python
#파일에 적힌 순서대로 각 값들을 나열하기
n_rom = [t for t in preprocessed if t in roman] 
n_sub = [t for t in preprocessed if t in subtitle]
n_num = [t for t in preprocessed if t in num_dot]
```

```python
n_num #170개 챕터
```

```python
n_num_index = [] #챕터들의 '인덱스'를 담은 리스트 만들기
for p in preprocessed:
    if p in n_num:
        n_num_index.append(preprocessed.index(p))
```

```python
len(n_num_index) #챕터들의 인덱스를 담은 리스트의 길이
```

```python
preprocessed.index('1·3') #리스트에서 값에 대한 인덱스 반환
```

```python
' '.join(preprocessed[3+1:27]) #' '.join(preprocessed[3:27])
```

```python
#각 챕터에 대응되는 본문 내용을 리스트로 담음(챕터 개수와 마찬가지로 총 170개)
content = []
for idx in range(len(n_num_index)):
    if idx < len(n_num_index)-1:
        tex=' '.join(preprocessed[n_num_index[idx]+1:n_num_index[idx+1]])
        content.append(tex)
    else:
        tex=' '.join(preprocessed[n_num_index[idx]+1:])
        content.append(tex)
```

```python
dic_num_content={n:c for n, c in zip(n_num, content)} # {챕터명:내용} 형식의 딕셔너리 생성
```

```python
dic_num_content['14·38']
```

```python
dic_num_content.keys()
```

```python
#소제목('Liang Hui Wang II' 등)별로 '챕터'를 묶은 리스트를 담은 하나의 리스트 생성(이중 리스트)
ls = []
for i in range(1,15): 
    ls2 = []
    for n in n_num:
        if n.startswith(str(i)+'·'):
            ls2.append(n)
    ls.append(ls2)
```

```python
ls[:3]
```

```python
#소제목('Liang Hui Wang II' 등)별로 '본문 내용'을 묶은 리스트를 담은 하나의 리스트 생성(이중 리스트)
sl = []
for c in ls:
    sl2 = []
    for k in c:
        sl2.append(dic_num_content[k])
    sl.append(sl2)
```

```python
sl[1]
```

```python
#소제목별로 각 챕터를 key, 챕터에 대응되는 내용을 value로 하는 딕셔너리를 만들고 각 딕셔너리를 담은 하나의 리스트를 생성
new_ls=[]
for k, v in zip(ls, sl): #[['1·3'], ['2·3', '2·9', '2·13', '2·14', '2·15'], ...] , [[내용], [내용,내용,내용,내용,내용], ...]
    dic={}
    for x, y in zip(k,v):
        dic[x]=y
    new_ls.append(dic)
```

```python
type(new_ls)
```

```python
#소제목을 key로, 위에서 만든 딕셔너리를 value로 하는 딕셔너리를 만들고 그 딕셔너리들을 담은 하나의 리스트를 생성
#여기서 만들어진 리스트는 뒤에서 제목(로마자 숫자(I, II, III, IV, V, ...))에 대응됨
new_ls2=[{k:v} for k, v in zip(n_sub, new_ls)] 
```

```python
len(new_ls2)
```

```python
#제목을 key로, 위에서 만든 딕셔너리를 value로 하는 하나의 딕셔너리를 생성(최종 데이터 구조)
mencius_data_structure={k:v for k, v in zip(n_rom, new_ls2)} 
```
